```python
from google_play_scraper import app  # Import fungsi 'app' dari pustaka google_play_scraper untuk mengambil informasi aplikasi dari Google Play Store.
import pandas as pd  # Import pustaka pandas dengan alias 'pd' untuk manipulasi dan analisis data.
import numpy as np   # Import pustaka numpy dengan alias 'np' untuk operasi array dan komputasi numerik.
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.quipper.school.assignment',
    lang='id',      # bahasa ulasan yang diinginkan
    country='id',   # negara pengguna
    sort=Sort.MOST_RELEVANT,  # cara mengurutkan ulasan, bisa menggunakan Sort.MOST_RELEVANT atau Sort.NEWEST
    count=1000,     # jumlah maksimum ulasan yang ingin diambil
    filter_score_with=None  # filter untuk menampilkan ulasan dengan skor tertentu, None (semua skor)
)
```

```python
df = pd.DataFrame(np.array(result), columns=['review'])  # Membuat DataFrame dari hasil ulasan dengan kolom 'review'.
df = df.join(pd.DataFrame(df.pop('review').tolist()))  # Memecah nilai dalam kolom 'review' yang berupa list menjadi kolom-kolom terpisah dalam DataFrame.
df.head()  # Menampilkan lima baris pertama dari DataFrame yang telah dimodifikasi.
```

```python
len(df.index) #menghitung jumlah baris data
```

```python
df_ig = df[['at','score','content']] #menyimpan dataframe baru
df_ig.head() #menampilkan 5 data teratas pada dataframe
```

```python
df_ig = df[['at', 'score', 'content']]  # Memilih kolom 'at' (tanggal ulasan), 'score' (nilai ulasan), dan 'content' (isi ulasan) dari DataFrame df.
sorted_df = df_ig.sort_values(by='at', ascending=False)  # Mengurutkan DataFrame df_ig berdasarkan kolom 'at' secara menurun (dari yang terbaru).
sorted_df.head()  # Menampilkan lima baris pertama dari DataFrame yang telah diurutkan.
```

```python
sorted_df.to_csv("quipper_data.csv", index=False) #menyimpan data dalam bentuk excel
```
